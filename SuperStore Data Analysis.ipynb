```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
data = pd.read_csv("SampleSuperstore.csv")
print("Data imported successfully")

data.head()
```

```python
data.isnull().sum()
```

```python
data['Category'].value_counts()
```

```python
data['Sub-Category'].value_counts()
```

```python
#Lets see how sub-categories are distributed wrt to category
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=data,color='y')
plt.show()
```

```python
# Profit and Sales according to sub category
plt.figure(figsize=(16,8))

data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.legend('Profit')
plt.legend('Sales')
plt.show()
```

```python
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot()
plt.title('Sales with respect to Profit')
plt.legend('Profit')
plt.legend('Sales')
plt.show()
```

```python
# Profit and Sales according to category
data.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.legend('Profit')
plt.legend('Sales')
plt.show()
```

```python
# Distribution of sub category wise products
plt.figure(figsize=(12,10))
data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
# Distribution of Category wise products
plt.figure(figsize=(12,10))
data['Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
#Region wise sub category count
import seaborn as sns
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=data)
plt.show()
```

```python
#Region wise category's count
import seaborn as sns

plt.figure(figsize=(15,8))
sns.countplot(x="Category", hue="Region", data=data)
plt.show()
```

```python
fig=plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
s=sns.countplot('Segment', data = data)
for s in ax.patches:
    ax.annotate('{:.0f}'.format(s.get_height()), (s.get_x()+0.15, s.get_height()+1))
plt.show()
```
